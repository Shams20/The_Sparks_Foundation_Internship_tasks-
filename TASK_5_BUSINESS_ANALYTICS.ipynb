# Here we make the dummies variables

```python
df['Ship Mode']= pd.get_dummies(df["Ship Mode"],drop_first = True)    
df.Segment= pd.get_dummies(df["Segment"],drop_first = True) 
#df.Country = pd.get_dummies(df["Country"],drop_first = True)  
df.Region= pd.get_dummies(df["Region"],drop_first = True)  
df.Category= pd.get_dummies(df["Category"],drop_first = True)

 
df['Sub-Category'] = pd.get_dummies(df["Sub-Category"],drop_first=True)
```

```python
df.head()
```

```python
## X is predicted value and Y is targeted value###
X = df.drop(['Country', 'City','State','Postal Code'], axis = 1)
y = df['Postal Code']
```

# Split the dataset into train & test

```python
# splitting X and y into training and testing sets 
    
from sklearn.model_selection import train_test_split 


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0) 
```

```python
print('shape of x_train',X_train.shape)
print('Shape of y_train: ',y_train.shape)

print('\n')

print('Shape of X_test: ',X_test.shape)
print('Shape of y_test: ',y_test.shape)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

# Training Algorithm

```python
from sklearn.linear_model import LinearRegression

regressor=LinearRegression()
regressor.fit(X_train,y_train)  # Fitting the model

print("Training Complete")
```

```python
# To retrive intercept & Coefficient

print("intercept is :")
print(regressor.intercept_)
```

```python
print("coefficient is :")
print(regressor.coef_)
```

```python
print(X_test)   # training data

y_pred= regressor.predict(X_test)
```

```python
# comparing Actual v/s Predicted

df=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```

```python
plt.figure(figsize=(14,6))

df.plot(kind='bar', figsize=(12,6), cmap="Set1")

plt.title('Actual vs Predicted score', fontsize=22)  
plt.tight_layout()
```
